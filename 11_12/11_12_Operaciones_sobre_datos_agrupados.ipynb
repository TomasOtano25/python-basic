# 11.12 Operaciones sobre datos agrupados

```python
import numpy as np
import pandas as pd
```

```python
ratings = pd.read_csv(r'ratings.csv')

peliculas = pd.read_csv(r'peliculas.csv')

usuarios = pd.read_csv(r'usuarios.csv')
```

### Un rating requiere un usuario y una pelicula, y los DataFrames estan vinculados por una clave (en este caso la clava es user_id y peli_id).

### Es posible que un usuario este asociado con 0 o varios ratings y peliculas. De la misma forma, una pelicula puede tener muchos ratings o ninguno, por un numero de diferentes usuarios.

```python
peli_ratings = pd.merge(peliculas, ratings, on=['peli_id'])

clasif = pd.merge(peli_ratings, usuarios)
```

```python
mas_ratings = clasif.groupby('titulo').size().sort_values(ascending=False)[:25]
mas_ratings
```

## ***.agg()***

### Este metodo nos permite aplicar una lista de funciones en columnas especificas del `DataFramas`.

### Deseamos conocer el rating promedio de cada pelicula.

```python
peli_stats = clasif.groupby('titulo').agg({'rating': [np.size, np.mean]})

peli_stats.head(15)
```

```python
peli_stats.sort_values([('rating', 'mean')], ascending=False).head(15)
```

```python
minimo_100 = peli_stats['rating']['size'] >= 100
```

```python
peli_stats[minimo_100].sort_values([('rating', 'mean')], ascending=False)[:15]
```
