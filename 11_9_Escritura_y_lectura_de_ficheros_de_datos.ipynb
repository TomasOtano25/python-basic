## ***.to_csv()***

### Este metodo nos permite exportar el `DataFrame` a un csv.

```python
df.to_csv('datos.csv')
```

```python
# df2 = pd.read_csv('datos.csv')
df2 = pd.read_csv('datos.csv', index_col=0)
df2.head()
```

```python
# na_rep = valor representante de los nulos (valor que sustituye en Null y NaN)
df.to_csv('nuevos-datos.csv', header=True, na_rep='(missing)')
```

```python
ti_da = {'POP': 'float32', 'AREA': 'float32', 'GDP': 'float32'}

# parse_dates = columnas de tipo fecha
# dtype = diccionario con tipos de datos
df3 = pd.read_csv('datos.csv', index_col=0, parse_dates=['IND_DAY'], dtype=ti_da)

df3.dtypes
```

```python
df3.IND_DAY
```

```python
# date_format = el formato de la fechas
df3.to_csv('datos-fechas-format.csv', date_format='%B %d, %Y')
```

```python
df4 = pd.read_csv('datos-fechas-format.csv', index_col=0)
df4.head()
```

```python
#usecols = usar columnas
df5 = pd.read_csv('datos.csv', index_col=0, usecols=[0, 1, 3, 5])
df5.head()
```
