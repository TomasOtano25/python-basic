# 12.1 Figuras y paneles

## Matplotlib

### Hacer graficos y visualizaciones es una de las tareas mas importantes a la hora de procesar y analizar datos. Puede formar parte del analisi exploratorio de datos para identificar valores extremos (outliers) o como medio para comunicacion de resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# figura de la funcion seno
fig, panel = plt.subplots()

# linspace(valor de inicio, valor final, cantidad de valores generados)
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

panel.plot(x, y)
```

### Podemos crear una figura con varios paneles

```python
# Crear la figura con dos paneles horizontales
fig, (pan_1, pan_2) = plt.subplots(2, 1)

# Figura 1
x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

pan_1.plot(x1, y1)

# Figura 2
x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)\

pan_2.plot(x2, y2)

plt.savefig(r'sin_cos.png')
```

```python
# Crear la figura con dos paneles verticales
fig, (pan_1, pan_2) = plt.subplots(1, 2, constrained_layout=True)

# Figura 1
x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

pan_1.plot(x1, y1)

# Figura 2
x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)

pan_2.plot(x2, y2)

plt.savefig(r'sin_cos_vertical.png')
```

```python

fig, ((pan_1, pan_2), (pan_3, pan_4)) = plt.subplots(2, 2, constrained_layout=True)

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

pan_1.plot(x, y)

pan_2.plot(x, y**2)

pan_3.plot(x, -y)

pan_4.plot(x, -y**2)

for ax in fig.get_axes():
  ax.set(xlabel='x-label', ylabel='y-label')
```

### El parametro sharex nos permite indicar que los paneles comparten el eje X

### El parametro sharexy nos permite indicar que los paneles comparten el eje Y

```python
fig, (pan_1, pan_2) = plt.subplots(2, sharex=True)

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

pan_1.plot(x, y, 'tab:orange')

pan_2.plot(x + 1, -y, 'green')
```

```python
fig, (pan_1, pan_2) = plt.subplots(1, 2, sharey=True)

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

pan_1.plot(x, y)

pan_2.plot(-x, y+1, 'green')
```

```python
fig, paneles = plt.subplots(2, 2, sharex='col', sharey='row')

(pan_1, pan_2), (pan_3, pan_4) = paneles

pan_1.plot(x, y)

pan_2.plot(x, y**2, 'tab:orange')

pan_3.plot(x + 1, -y, 'tab:green')

pan_4.plot(x + 2, -y**2, 'tab:red')
```

## .twinx()

### Este metodo nos permite graficar ambas funciones en el mismo panel

```python
fig, pan_1 = plt.subplots(1, 1)

pan_2 = pan_1.twinx()

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

pan_1.plot(x1, y1, 'y')

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)

pan_1.plot(x2, y2, 'r')

plt.savefig(r'sin_cos_2.png')
```
