# 11.7 Manaje de datos ausentes

`pandas` tomas a los valores `NaN` y `None` como valores ausentes. La funcion `pandas.isnull()` se puede usar para determinar si existen valores ausentes en los datos.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'VarA': ['aa', None, 'cc'],
    'VarB': [20, 30, None],
    'VarC': [1234, 3456, 6789]
  },
  index=['Caso 1', 'Caso 2', 'Caso 3'])

df
```

```python
df.isnull()
```

### Debemos decidir si descartamos las filas que incluyen valores ausentes, esto lo conseguimos usando el metodo `.dropna()`:

```python
df.dropna(subset=['VarA', 'VarB'])
```

### Otra opcion es sustituir los valores `NaN` con un valor determinadao, para esto usamos el metodo `.fillna()`.

```python
df.fillna("")
```

```python
# Remplazar los valores nulos por la media aritmetica
df.fillna(df.mean())
```

```python
df.fillna(25)
```
