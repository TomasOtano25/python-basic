## ***pd.merge()***

```python
clima_p = pd.read_csv(r'ny_precipitaciones.csv')
clima_p.shape
```

```python
clima_p.head()
```

```python
clima_t = pd.read_csv('ny_temperaturas.csv')
clima_t.shape
```

```python
clima_p.NAME
```

```python
clima_p[clima_p["NAME"] == "ITHACA CORNELL UNIVERSITY, NY US"]
```

```python
precip_itaca = clima_p[clima_p["NAME"] == "ITHACA CORNELL UNIVERSITY, NY US"]
precip_itaca.shape
```

### De forma predeterminada `pd.merge` realiza un *Inner Join* entre ambos DataFrames

```python
itaca_inner_merge = pd.merge(precip_itaca, clima_t)
itaca_inner_merge.shape
```

```python
itaca_inner_merge.head()
```

### Podemos especificar a `pd.merge` que clase `Join` deseamos hacer entre los DataFrames

```python
# Outer join (selecciona los elementos comunes como los no comunes)

# how = como
itaca_outer_merge = pd.merge(precip_itaca, clima_t, how="outer", on=["STATION", "DATE"])
itaca_outer_merge.columns
```

```python
itaca_outer_merge.shape
```

```python
itaca_outer_merge.head(15)
```

```python
# Left Join

itaca_left_join = pd.merge(precip_itaca, clima_t, how="left", on=["STATION", "DATE"])
itaca_left_join.shape
```

```python
itaca_left_join.head()
```

```python
# Right Join

itaca_right_join = pd.merge(precip_itaca, clima_t, how="right", on=["STATION", "DATE"])
itaca_right_join.shape
```

```python
itaca_right_join.head(15)
```

## ***.join()***

### Este es un metodo contenido en la clase `DataFrame`.

### Se diferencia de `pd.merge()` en que aquel es una funcion del modulo pandas y que `.join()` (de forma predeterminada) busca unir los conjunto a partir de los indices.

```python
clima_t.join(clima_p)
```

```python
clima_join = clima_t.join(clima_p, lsuffix='_left')
clima_join.head()
```

```python
clima_join.shape
```

```python
clima_join.columns
```

```python
# mueve las columnas para convertirlas en indices
clima_p.set_index(["STATION", "DATE"])
```

```python
clima_joined_total = clima_t.join(clima_p.set_index(["STATION", "DATE"]),
                                  lsuffix="_x",
                                  rsuffix="_y",
                                  on=["STATION", "DATE"])

for i in clima_joined_total.columns:
  print(i)
```

```python
clima_joined_total.shape
```

## ***pd.concat()***

Esta funcion nos permite juntar DataFrames a traves de un eje, o las filas o las columnas.

```python
# agrega las columnas
clima_total_outer_concat = pd.concat([clima_t, clima_p], axis=1)
clima_total_outer_concat.shape
```

```python
clima_total_outer_concat.tail()
```

```python
# agrega las filas
clima_total_outer_concat = pd.concat([clima_t, clima_p], axis=0)
clima_total_outer_concat.shape
```

```python
df_jerar = pd.concat([clima_t, clima_p], keys=["temp", "precip"])
df_jerar
```
