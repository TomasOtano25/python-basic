*   Define el metodo Distancia, cuyos parametros sean otro par de corrdenadas y calcule la distancia entre los dos punto y la muestre por pantalla.

> PROTIP: En Python, la funcion raiz cuadrada se debe importar del modulo math y se llama sqrt()





```python
import math

class Punto:

  def __init__(self,x=0,y=0):
    self.x = x
    self.y = y
  
  def __str__(self):
    return '({}, {})'.format(self.x,self.y)

  def cuadrante(self):
    if self.x > 0 and self.y > 0:
      print(f'{self} pertenece al cuadrante I.')
    elif self.x < 0 and self.y > 0:
      print(f'El punto pertenece al cuadrante II.')
    elif self.x < 0 and self.y < 0:
      print(f'El punto pertenece al cuadrante III.')
    elif self.x > 0 and self.y < 0:
      print(f'El punto pertenece al cuadrante IV.')

    # Si Y == 0, el punto se ubica sobre el eje X
    elif self.x !=0 and self.y == 0:
      print(f'{self} se ubica sobre el eje X')
    
    # Si X == 0, el punto se ubica sobre el eje Y
    elif self.x == 0 and self.y != 0:
      print(f'{self} se ubica sobre el eje Y')
    
    # Si ambos son 0, el punto se ubica en el origen
    else:
      print(f'{self} se ubica en el origen')

  def vector(self, p):
    print(f'EL vector que une los puntos {self} y {p} es igual a ({p.x-self.x}, {p.y-self.y})')

  def distancia(self, p):
    mod = math.sqrt( (p.x - self.x)**2 + (p.y - self.y)**2 )
    print(f'La distancia entre los puntos {self} y {p} es {mod}')
  
```

## Crear una clase llamada *Rectangulo* con dos puntos (inicial y final) que formaran la diagonal del rectangulo.

*   AÃ±ade un metodo constructor para crear ambos puntos facilmente, si no se envian se crearan dos puntos en el origen por defecto.
*   AÃ±ade al rectangulo un metodo llamada Mostrar_base que muestre la base.
*   AÃ±ade al rectangulo un metodo llamada Mostrar_altura que muestre la altura.
*   AÃ±ade al rectangulo un metodo llamada Mostrar_area que muestre el area.



```python
class Rectangulo:

  def __init__(self, inicio=Punto(), final=Punto()):
    self.inicio = inicio
    self.final = final
    self.base = abs(self.final.x - self.inicio.x) # abs = valor absoluto
    self.altura = abs(self.final.y - self.inicio.y)
    self.area = self.base * self.altura

  def mostrar_base(self):
    print(f'La base del rectangulo es {self.base}')

  def mostrar_altura(self):
    print(f'La base del rectangulo es {self.altura}')

  def mostrar_area(self):
    print(f'La base del rectangulo es {self.area}')
```
