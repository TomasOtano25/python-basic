# 11.3 Operaciones con DataFrames



```python
import numpy as np
import pandas as pd
```

```python
d = [np.random.randint(50, size=(10))]
d
```

```python
df = pd.DataFrame(d)
df
```

```python
df = pd.DataFrame(d).T
df
```

## AÃ±adir una nueva columna

```python
df['nueva_col'] = 5
df
```

```python
df['experiencia'] = 5
df
```

```python
df['perdidas'] = [(i+2)*np.e for i in range(10)]
df
```

```python
df['columna'] = df['perdidas']*100
df
```

## Cambiar el nombre de una columna

```python
df.columns
```

```python
df.columns = ['codigo_id', 'aÃ±os_exp', 'indice', 'eficiencia', 'eficiencia_agregada']
df
```

```python
df['codigo_id']
```

## Modificar una columna

```python
df['indice'] = df['indice'] / 200

df
```

## Eliminar una columna o una fila

```python
del df['eficiencia']

df
```

```python
# 0 elimina las filas y 1 elimina las columnas
c = df.drop([2,3,5], axis=0)
c
```

```python
df
```

```python
# axis = 0 elimina las filas y axis = 1 elimina las columnas
# inplace = eliminar los datos del dataframe original
df.drop(['indice'], axis=1, inplace = True)

df.head()
```

```python
df
```
