# **12.4 Tipos de graficos**

```python
import numpy as np
import matplotlib.pyplot as plt
```

## ***Graficos de lineas***

```python
año = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
PIB = [300.2, 543.3, 1075.9, 2862.2, 5979.6, 10289.7, 14958.3]

plt.plot(año, PIB, color='green', marker='o', linestyle='solid')

plt.title('PIB Nominal')

plt.ylabel('Millones de €')
plt.xlabel('Año')
```

## ***Scatter plot*** (Grafico de puntos)

```python
x1 = [3, 5, 7]
y1 = [3, 5, 9]

x2 = [1, 2, 3, 4]
y2 = [2, 3, 3, 2]

x3 = [8, 7, 2, 5, 4]
y3 = [6, 3, 7, 8, 7]

plt.scatter(x1, y1)

plt.scatter(x2, y2, marker='v', color='r')

plt.scatter(x3, y3, marker='^', color='m')

plt.title('Ejemplo de Scatter Plot')

plt.show()
```

## ***Graficos de barras***

```python
x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 7, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

plt.bar(x1, y1, label='Barra azul', color='b')

plt.bar(x2, y2, label='Barra verde', color='g')

plt.plot()

plt.xlabel('Nº barras')
plt.ylabel('Altura barras')
plt.title('Ejemplo de grafico de barras')
plt.legend('best')

plt.show()
```

## ***Histogramas***

```python
y = 5 + np.random.randn(100)

x = [e for e in range(len(y))]

plt.bar(x, y)

plt.title('Datos en crudo')
plt.show()
```

```python
plt.hist(y, bins=20)

plt.title('Histograma')
plt.show()
```

```python
plt.hist(y, cumulative=True, bins=20)

plt.title('Histograma acumulado')
plt.show()
```

## ***Graficos apilados***

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 1, 2, 3, 5]
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 9]

etiquetas = ["Sector privado", "Sector publico", 'Sector informal']

fig, panel = plt.subplots()

panel.stackplot(x, y1, y2, y3,
                labels= etiquetas,
                colors=['r', 'w', 'b'])

panel.legend(loc='upper left')

panel.set_title('Grafico apilado')

plt.show()
```

## ***Grafico de torta***

```python
etiquetas = 'Financiero', 'Salud', 'Construccion'
sectores = [56, 66, 24]
colores = ['c', 'b', 'r']

plt.pie(sectores, labels=etiquetas, colors=colores,
        startangle=90,
        explode=(0, 0.2, 0),
        autopct='%1.2f%%',
        shadow= True)

plt.axis('equal')
plt.title('Ejemplo de pie chart')
plt.show()
```

## ***Boxplot***



```python
# normal() = nos permite crear un arreglo de valores que pertenecer a una determinada media y determinada desviacion tipica
# normal(media, desviacion, numero de daos)
y1 = np.random.normal(100, 10, 100)
y2 = np.random.normal(80, 30, 100)
y3 = np.random.normal(90, 20, 100)
y4 = np.random.normal(70, 25, 100)

datos = [y1, y2, y3, y4]

fig, panel = plt.subplots()

panel.set_xticklabels(['Grupo_1', 'Grupo_2', 'Grupo_3', 'Grupo_4'])

bp = panel.boxplot(datos, patch_artist=True)

plt.title('Ejemplo de Boxplot')

# Elementos del boxplot

for box in bp['boxes']:
  box.set( color='r', linewidth=1 )
  box.set( facecolor= 'b')

for whisker in bp['whiskers']:
  whisker.set(color='g', linewidth=3)

for cap in bp['caps']:
  cap.set(color='black', linewidth=4)

for median in bp['medians']:
  median.set(color='#b2df8a', linewidth=2)

for flier in bp['fliers']:
  flier.set(marker='*', color='r', alpha=1)
```
